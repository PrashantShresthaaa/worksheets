```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/datasetsworkshop1/Copy of medical_students_dataset.csv')
print("Dataset Loaded Successfully!")
```

```python
#2.2
print("\nDataFrame Info:")
df.info()
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])

```

```python
#2.3
numerical_columns = ['Student ID', 'Age', 'Height', 'Weight', 'BMI', 'Temperature', 'Heart Rate',
                     'Blood Pressure', 'Cholesterol']

for col in numerical_columns:
    df[col].fillna(df[col].median(), inplace=True)
    print(f"Missing values in '{col}' column filled with median.")


categorical_columns = ['Gender', 'Blood Type', 'Diabetes', 'Smoking']

for col in categorical_columns:
    df[col].fillna(df[col].mode()[0], inplace=True)
    print(f"Missing values in '{col}' column filled with mode.")

```

```python

duplicates = df.duplicated().sum()
print(f"\nTotal Duplicate Rows are: {duplicates}")
if duplicates > 0:
    df = df.drop_duplicates()
    print(f"\nDuplicate rows were removed: {df.shape}")
else:
    print("\nNo duplicatess.")
```

```python
import pandas as pd

#

df = pd.read_csv('/content/drive/MyDrive/datasetsworkshop1/Copy of Copy of Titanic-Dataset.csv')

df_first_class = df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
df_first_class = df_first_class[df_first_class['Pclass'] == 1]

fare_mean = df_first_class['Fare'].mean()
fare_median = df_first_class['Fare'].median()
fare_max = df_first_class['Fare'].max()
fare_min = df_first_class['Fare'].min()

print(f"Mean of 'Fare': {fare_mean}")
print(f"Median of 'Fare': {fare_median}")
print(f"Maximum of 'Fare': {fare_max}")
print(f"Minimum of 'Fare': {fare_min}")

```

```python
#prob 2
null_age_count = df_first_class['Age'].isnull().sum()
print(f"Number of null values in 'Age': {null_age_count}")

df_first_class_cleaned = df_first_class.dropna(subset=['Age'])
print(f"\nDataFrame shape after dropping nulls: {df_first_class_cleaned.shape}")
```

```python
#prob 3
df_encoded = pd.get_dummies(df, columns=['Embarked'], prefix='Embarked')

print("\nFirst few rows after one-hot encoding:")
df_encoded.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
mean_survival_by_gender = df.groupby('Sex')['Survived'].mean()
print("Mean Survival Rates by Gender:")
print(mean_survival_by_gender)


plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=df, ci=None)

plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Gender on the Titanic')
plt.show()
```

```python


mean_survival_by_gender_embarked = df.groupby(['Sex', 'Embarked'])['Survived'].mean().reset_index()

#
plt.figure(figsize=(10, 6))
sns.barplot(x='Sex', y='Survived', hue='Embarked', data=mean_survival_by_gender_embarked)

plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.title('Survival Rate by Gender and Port of Embarkation on the Titanic')
plt.show()
```
